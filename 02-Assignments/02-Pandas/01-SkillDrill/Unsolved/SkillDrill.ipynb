```python
# Import dependencies
import pandas as pd
```

```python
# Read file into DataFrame
df = pd.read_csv('../Resources/ratings_and_sentiments.csv')
df.head()
```

```python
# Find all reviews with a number rating greater than or equal to 4.
df[df['num_rating']>=4]['review_text']
```

```python
# Find all the coffee shops with a overall sentiment greater than or equal to 4.
df[df['overall_sent']>=4]['coffee_shop_name']
```

```python
# Return the count of the above query.
df[df['overall_sent']>=4]['coffee_shop_name'].count()
```

```python
# Change the name of the review_text column to review
df.rename(columns={'review_text':'review'}, inplace=True)
df.head()
```

```python
# Change the name of the bool_HIGH column to bool_high
df.rename(columns={'bool_HIGH':'bool_high'}, inplace=True)
df.head()
```

```python
# Save your changes to the file as a pipe delimited file.
df.to_csv('../Resources/ratings_and_sentiments_pipe.csv',sep='|')
```
